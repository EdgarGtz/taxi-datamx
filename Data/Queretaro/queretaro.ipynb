```python
import pandas as pd
import numpy as np
```

```python
# Use tabula to read pdf files.
# http://tabula.technology/

# Upload converted csv file.
queretaro = pd.read_csv('queretaro_tabula_2017.csv')

# Select entries referring to taxis.
queretaro = queretaro[queretaro['Modalidad'] == 'TAXI'] 

# Create an ID column.
queretaro['id'] = queretaro.index 
queretaro.id += 1

#queretaro.index = np.arange(0, len(queretaro))
#queretaro['id'] = queretaro.index
#queretaro.id += 1

#queretaro = queretaro.reset_index()

# Delete unnecessary columns.
del queretaro['Modalidad']
del queretaro['NO\rCONSEC.']

# Rename columns.
queretaro.columns = ['concesion', 'nueva_concesion', 'zona', 'municipio', 'nombre', 'ultima_referencia', 'id']

# Reorder columns.
queretaro = queretaro[['id', 'nombre', 'municipio', 'zona', 'concesion', 'nueva_concesion', 'ultima_referencia']]

# Fill missing values.
#queretaro.nombre = queretaro.nombre.fillna('NA')
#queretaro.municipio = queretaro.municipio.fillna('NA')
#queretaro.zona = queretaro.zona.fillna('NA')
#queretaro.nueva_concesion = queretaro.nueva_concesion.fillna('NA')
queretaro.ultima_referencia = queretaro.ultima_referencia.fillna(0)

# Convert data types.
queretaro.concesion = queretaro.concesion.astype(int)
queretaro.ultima_referencia = queretaro.ultima_referencia.astype(int)
queretaro.ultima_referencia = pd.to_datetime(queretaro.ultima_referencia, format = '%Y', errors = 'coerce')

# Convert text values to lower case.
queretaro.nombre = queretaro.nombre.str.lower()
queretaro.municipio = queretaro.municipio.str.lower()
queretaro.zona = queretaro.zona.str.lower()

# Remove carriage returns.
queretaro.nombre = queretaro.nombre.str.replace(r'\r', ' ')
queretaro.nombre = queretaro.nombre.str.replace(r'\|\|', '-')

# Export to a csv file. 
#queretaro.to_csv('queretaro.csv')
```
